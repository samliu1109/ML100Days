2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?

Ans:可以，限制參數w的增長，使w不會變得特别大，那麼模型對輸入w中噪聲的敏感度就會降低，這就是Ridge Regression的作法，以此可以處理共線性問題。
