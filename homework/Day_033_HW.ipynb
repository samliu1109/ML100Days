### 1. 模型的泛化能力 (generalization) 是指什麼？ 

Ans:是指機器學習對於新鮮樣本的適應能力，也就是說是為學習到隱藏在數據背後的規律，經過訓練的樣本會對新樣本作出適合的預測，這就是泛化能力。
### 2. 分類問題與回歸問題分別可用的目標函數有哪些？

Ans:2. 回歸常用的目標函數: 均方誤差(Mean square error，MSE)和平均絕對值誤差(Mean absolute error，MAE)
       分類問題常用的目標函數: 交叉熵(cross-entropy)。
